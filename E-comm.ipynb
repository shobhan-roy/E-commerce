```python
import datetime
new_date= datetime.datetime(2017,12,20)
new_date
```

```python
RFMApproach = df1.groupby('CustomerID').agg({'Date': lambda x:(new_date - x.max()).days,
                                             'Invoice_no' : lambda x:x.count(),
                                             'Price': lambda x: x.sum()})
RFMApproach
```

```python
RFMApproach.rename(columns= {'Date': 'Recency', 'Invoice_no': 'Frequency', 'Price': 'Monetary'}, inplace= True)
RFMApproach.reset_index()
```

```python
RFMApproach.describe()
```

```python
quantiles= RFMApproach.quantile(q=[0.25, 0.5, 0.75])
quantiles= quantiles.to_dict()
quantiles
```

```python
def recency_value (x,d,p):
  if x<=p[d][0.25]:
    return 1
  elif x<=p[d][0.5]:
    return 2
  elif x<=p[d][0.75]:
    return 3
  else:
    return 4
```

```python
def freq_mont_value (x,d,p):
  if x<=p[d][0.25]:
    return 4
  elif x<=p[d][0.5]:
    return 3
  elif x<=p[d][0.75]:
    return 2
  else:
    return 1
```

```python
RFMApproach['R']= RFMApproach.Recency.apply(recency_value, args=['Recency', quantiles])
RFMApproach['F']= RFMApproach.Frequency.apply(freq_mont_value, args=['Frequency', quantiles])
RFMApproach['M']= RFMApproach.Monetary.apply(freq_mont_value, args=['Monetary', quantiles])
RFMApproach.reset_index()
```

```python
RFMApproach['RFM']= RFMApproach[['R','M','F']].sum(axis=1)
```

```python
RFMApproach
```

```python
RFMApproach.RFM.unique()
```

```python
Loyality= ['Platinum', 'Gold', 'Silver', 'Bronze']
score= pd.qcut(RFMApproach.RFM, q=4, labels= Loyality)
RFMApproach['Loyality_level']= score.values
RFMApproach= RFMApproach.reset_index()
RFMApproach
```

```python
RFMApproach.to_csv('segmenation_customer_details.csv')
```

```python
final_report = RFMApproach.groupby('Loyality_level')[['Recency', 'Frequency', 'Monetary','CustomerID']].agg({
    'Recency': 'sum', 'Frequency':'sum', 'Monetary':'sum','CustomerID':'nunique'}).reset_index()
```
